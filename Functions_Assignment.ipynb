1. What is the difference between a function and a method in Python?

- A function is an independent block of code that performs a task.
A method is a function that is associated with an object and called using the object.

Example:

```python
# Function Example
def add(a, b):
    return a + b

print(add(2, 3))  # Output: 5

# Method Example
class MyClass:
    def display(self):
        print("This is a method.")

obj = MyClass()
obj.display()  # Output: This is a method.
```

10. What is the difference between map(), reduce(), and filter() functions in Python?

- map(): Applies a function to each element.
- filter(): Filters elements based on a condition.
- reduce(): Reduces an iterable to a single value.

Example:

```python
from functools import reduce

nums = [1, 2, 3, 4]
print(list(map(lambda x: x**2, nums)))  # [1, 4, 9, 16]
print(list(filter(lambda x: x % 2 == 0, nums)))  # [2, 4]
print(reduce(lambda x, y: x + y, nums))  # 10
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_even(nums):
    return sum(num for num in nums if num % 2 == 0)

print(sum_even([1, 2, 3, 4]))  # Output: 6
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print([n for n in range(1, 201) if is_prime(n)])
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class Fibonacci:
    def __init__(self, terms):
        self.a, self.b = 0, 1
        self.terms = terms

    def __iter__(self):
        return self

    def __next__(self):
        if self.terms <= 0:
            raise StopIteration
        self.terms -= 1
        self.a, self.b = self.b, self.a + self.b
        return self.a

fib = Fibonacci(5)
print(list(fib))  # Output: [1, 1, 2, 3, 5]
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_2(exp):
    for i in range(exp + 1):
        yield 2**i

print(list(powers_of_2(5)))  # Output: [1, 2, 4, 8, 16, 32]
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_lines(file):
    with open(file, "r") as f:
        for line in f:
            yield line.strip()
```

11. Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:

 order number   . book title and author                          .Quantity          .price per item

34587                  learning python, mark lutz                   4                           40.95

98762                  programming python, mark lutz          5                           56.80

77226                  head first python, paul barry                3                           32.95

88112                  elinfurg in python, bernd kelin             3                           24.99

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100 €.

Write a Python program using lambda and map.

**Solution**

we have a task for creating a Python program that processes bookshop orders. Each order includes:

Order number

Book title and author

Quantity

Price per item

The goal is to return a list of 2-tuples, where each tuple contains:

- Order number

- The total price (quantity × price per item). If the total price is less than €100, add €10 to it.

```python
# Input: List of orders with order number, book details, quantity, and price per item
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python, Bernd Klein", 3, 24.99],
]

# Lambda function to calculate total price
calculate_total = lambda order: (
    order[0],  # Order number
    (order[2] * order[3] + 10) if (order[2] * order[3]) < 100 else (order[2] * order[3])
)

# Using map to apply the calculation to each order
result = list(map(calculate_total, orders))

# Output the result
print(result)
```

Explanation

1.Input:

The list orders contains sublists where:

- order[0]: Order number
- order[1]: Book title and author (not used for the calculation)
- order[2]: Quantity
- order[3]: Price per item

2.Logic:

- For each order, calculate the total price as quantity × price per item.
- If the total price is less than €100, add €10 to the total.
- A lambda function handles this logic.

3.map():

- The map function applies the lambda function to each order in the orders list.

4.Result:

A list of 2-tuples, where each tuple contains:

- Order number
- Adjusted total price
